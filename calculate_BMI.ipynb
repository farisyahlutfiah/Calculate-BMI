# 2. Hitung BMI

```python
# Mengubah tinggi badan cm ke meter
df_h['tinggi_badan_m'] = df_h['tinggi_badan'] / 100
df_h

# Hitung BMI, membulatkannya, dan update df_h
df_h['BMI'] = round(df_h['berat_badan'] / (df_h['tinggi_badan_m'] ** 2), 1)
df_h
```

# 3. Membuat program hitung BMI

```python
# Fungsi untuk menghitung BMI
def hitung_bmi(berat, tinggi):
    bmi = berat / (tinggi ** 2)
    return bmi

# Input berat badan (kg) dan tinggi badan (m)
berat = float(input("Masukkan berat badan Anda (kg): "))
tinggi = float(input("Masukkan tinggi badan Anda (m): "))

# Hitung BMI
bmi = hitung_bmi(berat, tinggi)

# Menampilkan hasil BMI
print("Indeks Massa Tubuh (BMI) Anda adalah:", round(bmi, 1))
```

# 9. Cari nilai-nilai berikut:

## Kelompokkan tinggi badan dan berat badan serta cari nilai terendah dan tertingginya

```python
# Kelompokkan berat badan dan tinggi badan
berat_kelompok = pd.cut(df_h['berat_badan'], bins=[0, 50, 60, 70, 80, float('inf')],
                        labels=['<50', '50-60', '60-70', '70-80', '>80'])
tinggi_kelompok = pd.cut(df_h['tinggi_badan'], bins=[0, 160, 170, 180, 190, float('inf')],
                         labels=['<160', '160-170', '170-180', '180-190', '>190'])

# Temukan nilai terendah dan tertinggi
berat_terendah = df_h['berat_badan'].min()
berat_tertinggi = df_h['berat_badan'].max()
tinggi_terendah = df_h['tinggi_badan'].min()
tinggi_tertinggi = df_h['tinggi_badan'].max()

print("Kelompok Berat Badan:")
print(berat_kelompok.value_counts())
print("\nKelompok Tinggi Badan:")
print(tinggi_kelompok.value_counts())
print("\nBerat Badan Terendah:", berat_terendah)
print("Berat Badan Tertinggi:", berat_tertinggi)
print("Tinggi Badan Terendah:", tinggi_terendah)
print("Tinggi Badan Tertinggi:", tinggi_tertinggi)
```
